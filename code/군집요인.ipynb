```python
import pandas as pd
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/군집.csv', encoding='utf-8')
df.set_index('부산광역시',inplace=True)
df.head()
```

```python
sns.pairplot(df, kind='reg')
plt.tight_layout()
```

```python
model = ols('pop ~ amount', data = df)
result = model.fit()
print(result.summary())
```

```python
model = ols('pop ~ house', data = df)
result = model.fit()
print(result.summary())
```

```python
model = ols('pop ~ trans', data = df)
result = model.fit()
print(result.summary())
```
